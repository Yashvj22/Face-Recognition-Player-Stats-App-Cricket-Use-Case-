## Single Function to Clean Data

+ Our Previous approach was lenghy, So we can make all Cleaning within a single function

```python
month_map = {
    "Jan": "01", "Feb": "02", "Mar": "03", "Apr": "04", "May": "05", "Jun": "06",
    "Jul": "07", "Aug": "08", "Sep": "09", "Oct": "10", "Nov": "11", "Dec": "12"
}
```

```python
def transform_value(value):
    value = str(value).strip()  # Convert to string and remove extra spaces
    
    if "-" in value:
        parts = value.split("-", 1)  # Split into two parts
        
        if parts[0] in month_map:  
            # Format: Month-Run (e.g., "May-61" -> "05/61")
            return f"{month_map[parts[0]]}/{parts[1]}"
        
        elif parts[1] in month_map:  
            # Format: Run-Month (e.g., "03-Oct" -> "10/03")
            return f"{month_map[parts[1]]}/{parts[0]}"
        
    return value
```

```python
data2 = data1.applymap(transform_value)
```

```python
l1 = list(data2.columns)
l1
```

```python
## Replacing Columns within our original Dataframe

for i in l1:
    data[i] = data2[i]
```
