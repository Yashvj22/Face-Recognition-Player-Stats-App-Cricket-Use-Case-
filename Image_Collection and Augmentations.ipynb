## Get only Face from Images using ML pre-trained Model

```python
import cv2
import os

# Load OpenCV's pre-trained face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

for folder in os.listdir("data/Images"):
    
    input_folder = os.path.join("data/Images", folder)
    output_folder = os.path.join(input_folder, "Faces")
    os.makedirs(output_folder, exist_ok=True)

    for c, image_name in enumerate(os.listdir(input_folder), start=1):
        
        img_path = os.path.join(input_folder, image_name)

        img = cv2.imread(img_path)

        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Detect faces
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

        if len(faces) > 0:
            for (x, y, w, h) in faces:
                face_crop = img[y:y+h, x:x+w]  # Crop the face
                face_resized = cv2.resize(face_crop, (64, 64))  # Resize to 64x64

                face_img_path = os.path.join(output_folder, f"{c}.jpg")
                cv2.imwrite(face_img_path, face_resized)

                print(f"[{folder}] Cropped and saved")
                break  # Save only first face
        else:
            print(f"[{folder}] No face detected in")
```

## Image Augmentation

```python

## Rotation of Images
def rotate_image(image, angle):
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)

    # Get rotation matrix
    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    # Apply the rotation
    rotated = cv2.warpAffine(image, M, (w, h), borderMode=cv2.BORDER_REFLECT)
    return rotated

## Scaling of Images
def scale_image(image, scale):
    (h, w) = image.shape[:2]
    scaled = cv2.resize(image, (int(w * scale), int(h * scale)))

    # Crop or pad to maintain 64x64 size
    if scale < 1.0:
        # Pad
        delta_w = 64 - scaled.shape[1]
        delta_h = 64 - scaled.shape[0]
        top, bottom = delta_h // 2, delta_h - delta_h // 2
        left, right = delta_w // 2, delta_w - delta_w // 2
        scaled = cv2.copyMakeBorder(scaled, top, bottom, left, right, cv2.BORDER_REFLECT)
    else:
        # Crop center
        start_x = (scaled.shape[1] - 64) // 2
        start_y = (scaled.shape[0] - 64) // 2
        scaled = scaled[start_y:start_y + 64, start_x:start_x + 64]

    return scaled

## Flipping of Images
def flip_image(image):
    return cv2.flip(image, 1) 

for player in os.listdir("data/Images"):
    face_folder = os.path.join("data/Images", player, "Faces")

    ## Sorting folder images to break loop for Augmented Images
    for img_name in sorted(os.listdir(face_folder)): 
        
        if "_rot" in img_name or "_scale" in img_name or "_flip" in img_name:
            break
            
        img_path = os.path.join(face_folder, img_name)
        img = cv2.imread(img_path)

        if img is None:
            continue

        base_name = os.path.splitext(img_name)[0]

        # Rotations
        for angle in [-30, -15, 15, 30]:
            rotated = rotate_image(img, angle)
            out_path = os.path.join(face_folder, f"{base_name}_rot{angle}.jpg")
            cv2.imwrite(out_path, rotated)

        # Scalings
        for scale in [0.8, 1.2]:
            scaled = scale_image(img, scale)
            out_path = os.path.join(face_folder, f"{base_name}_scale{scale}.jpg")
            cv2.imwrite(out_path, scaled)
            
        # Flipping
        flipped = flip_image(img)
        out_path = os.path.join(face_folder, f"{base_name}_flip.jpg")
        cv2.imwrite(out_path, flipped)
```

## Converting each Image into GrayScale to reduce Dimentionality

```python
for player in os.listdir("data/Images"):
    face_folder = os.path.join("data/Images", player, "Faces")

    for img_name in os.listdir(face_folder):
        img_path = os.path.join(face_folder, img_name)

        # Delete -15 and -30 rotation images
        if "_rot-15" in img_name or "_rot-30" in img_name:
            print(f"Deleting: {img_name}")
            os.remove(img_path)
        else:
            # Convert remaining image to grayscale
            img = cv2.imread(img_path)
            if img is None:
                continue
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            cv2.imwrite(img_path, gray)  # Overwrite with grayscale
            print(f"Converted to grayscale: {img_name}")
```

## Converting into DataFrame

```python
features = []
class_labels = []

for player in os.listdir("data/Images"):
    face_folder = os.path.join("data/Images", player, "Faces")

    for img_name in os.listdir(face_folder):
        img_path = os.path.join(face_folder, img_name)
        
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Read in grayscale directly even though images are in graysccale
        if img is None:
            continue

        img = img.flatten()
        features.append(img)
        class_labels.append(player)
```

```python
df = pd.DataFrame(features)
```

```python
df["Player"] = class_labels
```
